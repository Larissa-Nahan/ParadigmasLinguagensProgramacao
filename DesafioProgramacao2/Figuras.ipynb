```python
class Figura:
    def __init__(self):
        self

class Calcular:
    def calcular_perimetro(self, figuras):
        print("Perímetro")
        for i, figura in enumerate(figuras):
            print(f'==> O perímetro da figura {i+1} é {figura.calcular_perimetro()} cm')
    
    def calcular_area(self, figuras):
        print("\nÁrea")
        for i, figura in enumerate(figuras):
            print(f'==> A área da figura {i+1} é {figura.calcular_area()} cm²')
  
    def calcular_volume(self, figuras):
        print("\nVolume")
        for i, figura in enumerate(figuras):
            print(f'==> O volume da figura {i+1} é {figura.calcular_volume()} cm³')
            

            
class Quadrado(Figura):
    def __init__(self, altura, largura):
        super().__init__()
        self.altura = altura
        self.largura = largura
       
    def calcular_perimetro(self):
        return self.altura * 2 + self.largura * 2
    def calcular_area(self):
        return self.altura * self.largura

class Retangulo(Quadrado):
    def __init__(self, altura, largura):
        super().__init__(altura, largura)
       
    def calcular_perimetro(self):
        return super().calcular_perimetro()
    def calcular_area(self):
        return super().calcular_area()
    
class Circulo(Figura):
    def __init__(self, raio):
        super().__init__()
        self.raio = raio
    
    def calcular_perimetro(self):
        return 3.14 * self.raio * 2
    def calcular_area(self):
        return round(3.14 * self.raio * self.raio, 2)

class Triangulo(Figura):
    def __init__(self, altura, base):
        super().__init__()
        self.altura = altura
        self.base = base
        
    def calcular_perimetro(self):
        return self.base * 3
    def calcular_area(self):
        return self.altura * self.base * 1/2
    
class Trapezio(Figura):
    def __init__(self, base_menor, base_maior, altura, lado_1, lado_2):
        super().__init__()
        self.base_menor = base_menor
        self.base_maior = base_maior
        self.altura = altura
        self.lado_1 = lado_1
        self.lado_2 = lado_2
        
    def calcular_perimetro(self):
        return self.base_menor + self.base_maior + self.lado_1 + self.lado_2
    def calcular_area(self):
        return (self.base_menor + self.base_maior) * self.altura * 1/2
    
class Losango(Figura):
    def __init__(self, diagonal_menor, diagonal_maior, lado):
        super().__init__()
        self.diagonal_menor = diagonal_menor
        self.diagonal_maior = diagonal_maior
        self.lado = lado
        
    def calcular_perimetro(self):
        return self.lado * 4
    def calcular_area(self):
        return (self.diagonal_menor + self.diagonal_maior) * 1/2
    

class Cubo(Figura):
    def __init__(self, lado):
        super().__init__()
        self.lado = lado
        
    def calcular_volume(self):
        return self.lado * self.lado * self.lado
    
class Esfera(Figura):
    def __init__(self, raio):
        super().__init__()
        self.raio = raio
        
    def calcular_volume(self):
        return round(3.14 * self.raio * self.raio * self.raio * 4/3, 2)
    
class Piramide(Figura):
    def __init__(self, altura, lado):
        super().__init__()
        self.altura = altura
        self.lado = lado
        
    def calcular_volume(self):
        return round(self.lado * self.lado * self.altura * 1/3, 2)
    
class Cilindro(Figura):
    def __init__(self, altura, raio):
        super().__init__()
        self.altura = altura
        self.raio = raio
    def calcular_volume(self):
        return round(3.14 * self.raio * self.raio * self.altura, 2)
        
class Cone(Cilindro):
    def __init__(self, altura, raio):
        super().__init__(altura, raio)
        
    def calcular_volume(self):
        cilindro = super().calcular_volume()
        return round(cilindro * 1/3, 2)
```
